```python
import pandas as pd

data = pd.read_csv("superstore.csv", encoding="latin1")

data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.drop_duplicates()
```

```python
data.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.histplot(data['Sales'], bins=20, kde=True)
plt.title("Distribution of Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Profit', data=data)
plt.title("Profit by Category")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_data = data.select_dtypes(include='number')

plt.figure(figsize=(6,4))
sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Columns Only)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['Sales'], bins=20, kde=True)
plt.title("Distribution of Sales")
plt.show()
```

```python
sns.boxplot(x=data['Sales'])
plt.title("Boxplot of Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Region', data=data)
plt.title("Number of Orders by Region")
plt.show()
```

```python
import numpy as np

# randomly assign regions
regions = ['East', 'West', 'South', 'Central']
np.random.seed(42)
data['Region'] = np.random.choice(regions, size=len(data))

# now the countplot will work
plt.figure(figsize=(8,5))
sns.countplot(x='Region', data=data)
plt.title("Number of Orders by Region")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='SubCategory', data=data)
plt.title("Number of Orders by SubCategory")
plt.xticks(rotation=45)
plt.show()
```

```python
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
# Drop duplicates
data = data.drop_duplicates()
```
